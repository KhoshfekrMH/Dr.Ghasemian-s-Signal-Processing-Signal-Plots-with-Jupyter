```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
- What is a unit time step signal? [Unit Time Step Signal-Wikipedia](https://en.wikipedia.org/wiki/Unit_step_function)
- Range: t ∈ [-2, 100], x ∈ [-5, 5] & y ∈ [-2, 2]
  $$\begin{equation} u(t)  =\begin{cases}1, & \text{if } t \geq 0 \\ 0, & \text{if } t < 0\end{cases}\end{equation}$$

```python
time = np.linspace(-2, 100, 10000)
u_t = np.heaviside(time, 0)

plt.plot(time, u_t, 'r', linewidth=2, label='u(t)')
plt.xlim(-5, 5)
plt.ylim(-2, 2)

plt.title('Unit Time Step Signal')
plt.xlabel('t')
plt.ylabel('u(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```
