```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
- What is a unit time step signal? [Unit Time Step Signal-Wikipedia](https://en.wikipedia.org/wiki/Unit_step_function)
- Range: n ∈ [-2, 2], x ∈ [-3, 4] & y ∈ [-2, 2]
  $$\begin{equation} u[n]  =\begin{cases}1, & \text{if } n \geq 0 \\ 0, & \text{if } n < 0\end{cases}\end{equation}$$

```python
for n in range(-2, 2):
    n = n + 3
    if n < 0:
        y = 0    
    elif n >= 0:
        y = 1

    plt.stem(n, y, 'r', basefmt=' ')
    plt.xlim(-3, 4)
    plt.ylim(-2, 2)

    plt.title('Unit Time Step Discrete Signal')
    plt.xlabel('n')
    plt.ylabel('u[n]')
    plt.style.use('dark_background')
    plt.grid(True)
```
