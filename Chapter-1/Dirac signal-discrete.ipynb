```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
for  n in range(-2, 2):
    n = n + 1
    if n == 0:
        y = 1
    else:
        y = 0
        
    plt.stem(n, y, 'r', basefmt=' ')
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)

    plt.title('Dirac Delta Signal')
    plt.xlabel('n')
    plt.ylabel('δ[n]')
    plt.grid(True)    
```
