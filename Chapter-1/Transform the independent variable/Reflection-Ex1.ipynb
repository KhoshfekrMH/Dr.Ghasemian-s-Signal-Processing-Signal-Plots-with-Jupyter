```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
- In this file I will show how to plot reflected and standard signals
- For instance, I used $y = \sin(2\pi t)$ and $y = \sin(-2\pi t)$ to show how reflection works
- Also I little bit limit signals to range of t ∈ [0, 1]

### Standard Signal
$y = \sin(2\pi t)$

```python
A = 1
f = 1
phi = 0
sr = 100

time = np.arange(0, 1, 1/sr)
y = A*np.sin(2*np.pi*f*time + phi)
plt.figure(3)
plt.plot(time, y, color='r', label='sin(2πt)')

plt.title("Standard Sine Wave Signal")
plt.xlabel("time")
plt.ylabel("x1(t)")
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

### Reflected Signal
$y = \sin(-2\pi t)$

```python
A = 1
f = 1
phi = 0
sr = 100

time = np.arange(0, 1, 1/sr)
y = A*np.sin(-2*np.pi*f*time + phi) #As you see I multiply -2 instead of 2 and it cause reflection in signal
plt.figure(3)
plt.plot(time, y, color='g', label='sin(-2πt)')

plt.title("Reflected Sine Wave Signal")
plt.xlabel("time")
plt.ylabel("x2(t)")
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

### Comparison

```python
A = 1
f = 1
phi = 0
sr = 100

time = np.arange(0, 1, 1/sr)
y1 = A*np.sin(2*np.pi*f*time + phi)
y2 = A*np.sin(-2*np.pi*f*time + phi)
plt.figure(3)
plt.plot(time, y1, color='r', label='sin(2πt)')
plt.plot(time, y2, color='g', label='sin(-2πt)')

plt.title("Compare Two Sine Wave Signals")
plt.xlabel("time")
plt.ylabel("y")
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```
