```python
import numpy as np
import matplotlib.pyplot as plt
```

### Standard Signal

```python
A = 1
f = 1
phi = 0
sr = 100

time = np.arange(0, 1, 1/sr)
y = A*np.sin(2*np.pi*f*time + phi)
plt.figure(3)
plt.plot(time, y, color='r')

plt.title("Sine Wave Signal")
plt.xlabel("time")
plt.ylabel("sin(t)")
plt.style.use('dark_background')
plt.grid(True)
```

### Reflected Signal

```python
A = 1
f = 1
phi = 0
sr = 100

time = np.arange(0, 1, 1/sr)
y = A*np.sin(-2*np.pi*f*time + phi) #As you see I multiply -2 instead of 2 and it cause reflection in signal
plt.figure(3)
plt.plot(time, y, color='g')

plt.title("Reflected Sine Wave Signal")
plt.xlabel("time")
plt.ylabel("sin(-t)")
plt.style.use('dark_background')
plt.grid(True)
```

### Comparison

```python
A = 1
f = 1
phi = 0
sr = 100

time = np.arange(0, 1, 1/sr)
y1 = A*np.sin(2*np.pi*f*time + phi)
y2 = A*np.sin(-2*np.pi*f*time + phi)
plt.figure(3)
plt.plot(time, y1, color='r')
plt.plot(time, y2, color='g')

plt.title("Compare two Sine Wave Signals")
plt.xlabel("time")
plt.ylabel("y")
plt.style.use('dark_background')
plt.grid(True)
```
