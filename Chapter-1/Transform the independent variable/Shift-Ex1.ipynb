```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
- In this notebook I will plot these three signals:
  1. Standard ABS signal: $\left\lvert t  \right\rvert$
  2. Shifted right ABS signal: $\left\lvert t-3  \right\rvert$
  3. Shifted left ABS signal: $\left\lvert t+3  \right\rvert$
- All signals are plotted on the same graph and also same range: t ∈ [-100, 100], x ∈ [-20, 20] and y ∈ [-2, 2]

### Standard Signal
$\left\lvert t  \right\rvert$

```python
t = np.linspace(-100, 100, 10000)
y = np.abs(t)
plt.plot(t, y, 'r', label='|t|')

plt.xlim(-20, 20)
plt.ylim(-2, 2)

plt.title('plot of |t|')
plt.xlabel('time')
plt.ylabel('x1(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

#### Shifted Right ABS Signal
$\left\lvert t-3  \right\rvert$

```python
t = np.linspace(-100, 100, 10000)
y = np.abs(t - 3) #As you see by subtracting 3 from t, the graph shifts to the right
plt.plot(t, y, 'g', label='|t-3|')

plt.xlim(-10, 10)
plt.ylim(-1, 1)

plt.title('Shifted Right ABS Signal')
plt.xlabel('time')
plt.ylabel('x2(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

#### Shifted Left ABS Signal
$\left\lvert t+3  \right\rvert$

```python
t = np.linspace(-100, 100, 10000)
y = np.abs(t + 3) #As you see by adding 3 to t, the graph shifts to the left
plt.plot(t, y, 'y', label='|t+3|')

plt.xlim(-10, 10)
plt.ylim(-1, 1)

plt.title('Shifted Left ABS Signal')
plt.xlabel('time')
plt.ylabel('x3(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

### Comparison

```python
t = np.linspace(-100, 100, 10000)
y = np.abs(t)
plt.plot(t, y, 'r', label='|t|')
plt.plot(t - 3, y, 'g', label='|t+3|')
plt.plot(t + 3, y, 'y', label='|t-3|')

plt.xlim(-10, 10)
plt.ylim(-1, 1)

plt.title('Compare Shifted Signals')
plt.xlabel('time')
plt.ylabel('y')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```
