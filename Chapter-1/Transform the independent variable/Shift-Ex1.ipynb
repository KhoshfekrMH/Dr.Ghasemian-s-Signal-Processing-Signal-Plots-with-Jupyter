```python
import numpy as np
import matplotlib.pyplot as plt
```

### Standard Signal

```python
t = np.linspace(-5, 5, 100)
y = np.abs(t)
plt.plot(t, y, 'r')

plt.xlim(-10, 10)
plt.ylim(-1, 1)

plt.title('plot of |t|')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.style.use('dark_background')
plt.grid(True)
```

#### Shifted Right Signal

```python
t = np.linspace(-5, 5, 100)
y = np.abs(t - 3) #As you see by subtracting 3 from t, the graph shifts to the right
plt.plot(t, y, 'g')

plt.xlim(-10, 10)
plt.ylim(-1, 1)

plt.title('plot of |t|')
plt.xlabel('time')
plt.ylabel('x(t - 3)')
plt.style.use('dark_background')
plt.grid(True)
```

#### Shifted Left Signal

```python
t = np.linspace(-5, 5, 100)
y = np.abs(t + 3) #As you see by adding 3 to t, the graph shifts to the left
plt.plot(t, y, 'y')

plt.xlim(-10, 10)
plt.ylim(-1, 1)

plt.title('plot of |t|')
plt.xlabel('time')
plt.ylabel('x(t + 3)')
plt.style.use('dark_background')
plt.grid(True)
```

### Comparison

```python
t = np.linspace(-5, 5, 100)
y = np.abs(t)
plt.plot(t, y, 'r')
plt.plot(t - 3, y, 'g')
plt.plot(t + 3, y, 'y')

plt.xlim(-10, 10)
plt.ylim(-1, 1)

plt.title('plot of |t|')
plt.xlabel('time')
plt.ylabel('y')
plt.style.use('dark_background')
plt.grid(True)
```
