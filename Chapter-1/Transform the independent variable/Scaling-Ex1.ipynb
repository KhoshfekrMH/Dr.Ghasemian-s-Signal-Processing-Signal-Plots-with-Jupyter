```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
- How scaling in signals works
- I used abs functions: $\left\lvert t \right\rvert \epsilon [-100, 100]$ , $\left\lvert 2t \right\rvert \epsilon [-100, 100]$ and $\left\lvert \frac{t}{2}  \right\rvert \epsilon [-100, 100]$
- Both of them are limited in x and y axis

### Standard Signal
$\left\lvert t \right\rvert$

```python
t = np.linspace(-100, 100, 10000)
y = np.abs(t) #normal signal
plt.plot(t, y, 'r', label='|t|')

plt.xlim(-2, 2)
plt.ylim(0, 1)

plt.title('Standard ABS Signal')
plt.xlabel('time')
plt.ylabel('x1(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

### Scaled Signal (shrinkage, multiplication)
$\left\lvert 2t \right\rvert$

```python
t = np.linspace(-100, 100, 10000)
y = np.abs(t*2) #by multiply time by 2 Signal will shrink
plt.plot(t, y, 'y', label='|2t|')

plt.xlim(-2, 2)
plt.ylim(0, 1)

plt.title('Shrinked ABS Signal')
plt.xlabel('time')
plt.ylabel('x2(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

### Scaled Signal (expansion, Division)
$\left\lvert \frac{t}{2}  \right\rvert$

```python
t = np.linspace(-100, 100, 10000)
y = np.abs(t/2) #by divided time by 2 Signal will expanded
plt.plot(t, y, 'g', label="|t/2|")

plt.xlim(-2, 2)
plt.ylim(0, 1)

plt.title('Expanded ABS Signal')
plt.xlabel('time')
plt.ylabel('x3(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

### Comparison

```python
t = np.linspace(-100, 100, 10000)
y = np.abs(t)
#compare scaled signals
plt.plot(t, y, 'r', label='|t|')
plt.plot(t, y*2, 'y', label='|2t|')
plt.plot(t, y/2, 'g', label='|t/2|')

plt.xlim(-2, 2)
plt.ylim(0, 1)

plt.title('Compare Scaled Signals')
plt.xlabel('time')
plt.ylabel('y')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```
