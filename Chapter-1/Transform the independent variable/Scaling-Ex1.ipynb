```python
import numpy as np
import matplotlib.pyplot as plt
```

### Standard Signal

```python
t = np.linspace(-5, 5, 100)
y = np.abs(t) #normal signal
plt.plot(t, y, 'r')

plt.xlim(-2, 2)
plt.ylim(-1, 1)

plt.title('plot of x(t) = |t|')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.style.use('dark_background')
plt.grid(True)
```

### Scaled Signal (shrinkage, multiplication)

```python
t = np.linspace(-5, 5, 100)
y = np.abs(t*2) #by multiply time by 2 Signal will shrink
plt.plot(t, y, 'y')

plt.xlim(-2, 2)
plt.ylim(-1, 1)

plt.title('plot of x(t) = |2t|')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.style.use('dark_background')
plt.grid(True)
```

### Scaled Signal (expansion, Division)

```python
t = np.linspace(-5, 5, 100)
y = np.abs(t/2) #by divided time by 2 Signal will expanded
plt.plot(t, y, 'g')

plt.xlim(-2, 2)
plt.ylim(-1, 1)

plt.title('plot of x(t) = |t/2|')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.style.use('dark_background')
plt.grid(True)
```

### Comparison

```python
t = np.linspace(-5, 5, 100)
y = np.abs(t)
#compare scaled signals
plt.plot(t, y, 'r')
plt.plot(t, y*2, 'y')
plt.plot(t, y/2, 'g')

plt.xlim(-2, 2)
plt.ylim(-1, 1)

plt.title('plot of x(t) = |t/2|')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.style.use('dark_background')
plt.grid(True)
```
