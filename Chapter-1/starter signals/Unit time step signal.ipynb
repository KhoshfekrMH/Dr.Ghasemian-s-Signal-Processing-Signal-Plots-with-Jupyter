```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.linspace(-2, 100, 1000)
u_t = np.heaviside(time, 0)

plt.plot(time, u_t, 'r', linewidth=2)
plt.xlim(-5, 5)
plt.ylim(-2, 2)

plt.title('Unit Time Step Signal')
plt.xlabel('t')
plt.ylabel('u(t)')
plt.style.use('dark_background')
plt.grid(True)
```
