```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
- What is a ramp signal? [Ramp function-Wikipedia](https://en.wikipedia.org/wiki/Ramp_function)
- Range: t ∈ [-2, 100], x ∈ [-10, 10] & y ∈ [0, 10]
  $$\begin{equation} r(t)  =\begin{cases}t, & \text{if } t \geq 0 \\ 0, & \text{if } t < 0\end{cases}\end{equation}$$

```python
t = np.linspace(-2, 100, 10000)
ramp = np.piecewise(t, [t < 0, t >= 0], [0, lambda t: t])
plt.plot(t, ramp, 'r', linewidth=2, label='r(t)')
plt.xlim(-10, 10)
plt.ylim(0, 10)

plt.title('Ramp Signal')
plt.xlabel('time')
plt.ylabel('r(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```
