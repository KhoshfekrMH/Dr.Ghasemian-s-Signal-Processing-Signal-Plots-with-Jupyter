```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-2, 100, 1000)
ramp = np.piecewise(t, [t < 0, t >= 0], [0, lambda t: t])
plt.plot(t, ramp, 'r', linewidth=2)
plt.xlim(-10, 10)
plt.ylim(0, 10)

plt.title('Ramp Signal')
plt.xlabel('Ramp(t)')
plt.ylabel('time')
plt.style.use('dark_background')
plt.grid(True)
```
