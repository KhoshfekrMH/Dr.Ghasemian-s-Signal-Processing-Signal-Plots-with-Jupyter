```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
- What is Dirac Signal? [Dirac delta function-Wikipedia](https://en.wikipedia.org/wiki/Dirac_delta_function)
- Range: t ∈ [-50, 50] & y ∈ [0, 1]

$$\int_{\infty }^{-\infty } f(x)\delta (x) \,dx = f(0) \int_{\infty }^{-\infty } \delta (x) \,dx = f(0) $$

```python
t = np.linspace(-50, 50, 10000) 
delta = lambda t: np.exp(-t**2) # Define the Dirac delta function approximation

plt.axvline(x=0, color='red', linestyle='--', label='δ(t)') # δ(t)
plt.plot(t, delta(t), label='a -> 0')

plt.ylim(0, 1)

plt.title('Dirac Signal')
plt.xlabel('time')
plt.ylabel('δ(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```
