```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
- What is a sinusoidal signal or sine wave? [Sine wave- Wikipedia](https://en.wikipedia.org/wiki/Sine_wave)
- Range: t ∈ [0, 2] and A ∈ [0, 2]
$$y = A \sin(2\pi f t + \phi)$$
- I used these conditions to plot a sine wave signal:
  1. $A$ = 2 (peak amplitude)
  2. $f$ = 1 (frequency)
  3. $\phi$ = 0 (phase shift)
  4. sr = 100 (sampling rate) $\longrightarrow$ It is not exist in formula but in plot we need it
  5. So I plot this signal: $y = 2 \cdot \sin(2\pi t)$
  6. figure(3)

```python
A = 2
f = 1
phi = 0
sr = 100

time = np.arange(0, 2, 1/sr)
y = A * np.sin(2*np.pi*f*time + phi) # 2sin(2Πt)
plt.figure(3)
plt.plot(time, y, color="r", label="y=2sin(2Πt)")

plt.title("Sine Wave Signal")
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```
