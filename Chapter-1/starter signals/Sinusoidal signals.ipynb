```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 2
f = 1
phi = 0
sr = 100

time = np.arange(0, 2, 1/sr)
y = A * np.sin(2*np.pi*f*time + phi)
plt.figure(3)
plt.plot(time, y, color="r")

plt.title("Sine Wave Signal")
plt.xlabel("time")
plt.ylabel("sin(t)")
plt.style.use('dark_background')
plt.grid(True)
```
