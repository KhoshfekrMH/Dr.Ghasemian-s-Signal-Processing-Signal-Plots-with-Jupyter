```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-10, 10)  # time array from 0 to 1
y = np.exp(0.2*t)
plt.figure(4)
plt.plot(t, y, color="r")

plt.title('Real Exponential Signal')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.style.use('dark_background')
plt.grid(True)
```
