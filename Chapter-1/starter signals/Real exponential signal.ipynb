```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
- What is a real exponential signal? [Exponential function-Wikipedia](https://en.wikipedia.org/wiki/Exponential_function)
- Range: t âˆˆ [-10, 10], figure(4)
  $$a:Initial, b:Multiplier: Y = a.{b}^x$$
- I used $e^{0.2t}$ to show how this type of signal looks like because exponential signals vary in amplitude and shape

```python
t = np.linspace(-10, 10)  # time array from 0 to 1
y = np.exp(0.2*t) #e^(0.2t)
plt.figure(4)
plt.plot(t, y, color="r", label="e^(0.2t)")

plt.title('Real Exponential Signal')
plt.xlabel('time')
plt.ylabel('exp(0.2t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```
