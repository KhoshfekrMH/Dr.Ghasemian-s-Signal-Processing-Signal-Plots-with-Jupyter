```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
- This signal is made of discrete sinusoidal signal
- Range: n ∈ [-10, 10], x ∈ [-10, 10] & y ∈ [-2, 2]
$$x[n] = A \sin [2\pi f n + \phi]$$
- In this file I used $\sin[0.1πn]$ to show how discrete sinusoidal signals look like

```python
for n in range(-10, 10):
  plt.stem(n, np.sin(0.1*np.pi*n) , 'r', basefmt=' ')
  plt.xlim(-10, 10)
  plt.ylim(-2, 2)

  plt.title("Sine Wave Discrete Signal")
  plt.xlabel("n")
  plt.ylabel("sin[0.1πn]")
  plt.style.use('dark_background')
  plt.grid(True)
```
