```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
for n in range(-2, 2):
    n = n + 3
    if n < 0:
        y = 0    
    elif n >= 0:
        y = 1

    plt.stem(n, y, 'r', basefmt=' ')
    plt.xlim(-3, 4)
    plt.ylim(-2, 2)

    plt.title('Unit Time Step Signal')
    plt.xlabel('n')
    plt.ylabel('u[n]')
    plt.grid(True)

```
