```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
$$Signal: x(t) = \left\lvert t-2 \right\rvert u(t^2 - 1)$$
- Range: t ∈ [-10, 10], y ∈ [0, 6] and x ∈ [-4, 4]
- For plot this signal first of all, lets plot |t-2| and u(t^2 - 1) as separated signals

<hr>

#### Step 1(I): $\left\lvert t-2 \right\rvert$

```python
t = np.linspace(-10, 10, 10000)
x1 = np.abs(t-2)

plt.plot(t, x1, 'red', linewidth=3, label='|t-2|')

plt.xlim(-4, 14)
plt.ylim(0, 6)

plt.title('plot of |t|')
plt.xlabel('time')
plt.ylabel('x1(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

<hr>

#### Step 2(I): $u(t-1)$

```python
t = np.linspace(-10, 10, 10000)
f1 = np.heaviside(t - 1, 0) # u(t-2)

plt.plot(t, f1, 'aqua', linewidth=3, label='u(t-1)')

plt.xlim(-4, 4)
plt.ylim(0, 6)

plt.title('Step 2(I): u(t-1)')
plt.xlabel('time')
plt.ylabel('f1(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

#### Step 2(II): $u(t^2 - 1)$

```python
t = np.linspace(-10, 10, 10000)
x2 = np.heaviside(np.power(t, 2) - 1, 0) # u(t^2-1)

plt.plot(t, x2, 'aqua', linewidth=3, label='u(t^2-1)')

plt.xlim(-4, 4)
plt.ylim(0, 6)

plt.title('Step 2(II): u(t^2 - 1)')
plt.xlabel('time')
plt.ylabel('x2(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

<hr>

#### Step 3: $\left\lvert t-2 \right\rvert u(t^2 - 1)$

```python
t = np.linspace(-10, 10, 10000)
x1 = np.abs(t - 2) # |t-2|
x2 = np.heaviside(np.power(t, 2) - 1, 0) # u(t^2-1)
x_t = x1 * x2

plt.plot(t, x_t, 'orange', linewidth=3, label='|t-2|u(t^2 - 1)')

plt.xlim(-4, 4)
plt.ylim(0, 6)

plt.title('Step 3: |t-2|u(t^2 - 1)')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

<hr>

#### Comparison

```python
t = np.linspace(-10, 10, 10000)
# creation of x1:
x1 = np.abs(t - 2) # |t-2|
# creation of x2:
x2 = np.heaviside(np.power(t, 2) - 1, 0) # u(t^2-1)
# creation of |t-2|u(t^2 - 1):
x_t = x1 * x2

plt.plot(t, x1, 'red', linewidth=3, label='|t-2|')
plt.plot(t, x2, 'aqua', linewidth=3, label='u(t^2-1)')
plt.plot(t, x_t, 'orange', linewidth=6, alpha=0.5, label='|t-2|u(t^2 - 1)')

plt.xlim(-4, 4)
plt.ylim(0, 6)

plt.title('Compare Signals')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```
