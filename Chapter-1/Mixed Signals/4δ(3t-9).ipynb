```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
$$Signal: x(t) = 4\delta (3t-9)$$
- Range: t ∈ [-20, 20] and y ∈ [0, 1]
- In this file we will plot $4\delta (3t-9)$ in four steps:
1. $\delta (t)$
2. $\delta (t-9)$
3. $\delta (3t-9)$
4. $4\delta (3t-9)$

#### Step 1: $\delta (t)$

```python
t = np.linspace(-20, 20, 10000) 
delta = lambda t: np.exp(-t**2)

plt.axvline(x=0, color='red', linestyle='--', label='δ(t)', linewidth=2)
plt.plot(t, delta(t), label='a -> 0', alpha=0.6) # δ(t)

plt.ylim(0, 1)

plt.title('Signal δ(t)')
plt.xlabel('time')
plt.ylabel('x1(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

#### Step 2: $\delta (t-9)$

```python
t = np.linspace(-20, 20, 10000)
delta = lambda t: np.exp(-t**2)

plt.axvline(x=9, color='yellow', linestyle='--', label='δ(t-9)', linewidth=2)
plt.plot(t, delta(t - 9), label='a -> 0', alpha=0.6) # δ(t-9)

plt.ylim(0, 1)

plt.title('Signal δ(t-9)')
plt.xlabel('time')
plt.ylabel('x2(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

#### Step 3: $\delta (3t-9)$

```python
t = np.linspace(-20, 20, 10000)
delta = lambda t: np.exp(-t**2)

plt.axvline(x=3, color='Orange', linestyle='--', label='δ(3t-9)', linewidth=2)
plt.plot(t, delta(3*t - 9), label='a -> 0', alpha=0.6) # δ(3t-9)

plt.ylim(0, 1)

plt.title('Signal δ(3t-9)')
plt.xlabel('time')
plt.ylabel('x3(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

#### Step 4: 4δ(3t-9)

```python
t = np.linspace(-20, 20, 10000)
delta = lambda t: np.exp(-t**2)

plt.axvline(x=3, color='#50D890', linestyle='--', label='4δ(3t-9)', linewidth=2)
plt.plot(t, 4*delta(3*t - 9), label='a -> 0', alpha=0.6) # 4δ(3t-9)

plt.ylim(0, 1)

plt.title('Signal 4δ(3t-9)')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.legend()
plt.grid(True)
```

#### Comparison

```python
t = np.linspace(-5, 5, 1000)

plt.axvline(x=0, color='red', label='δ(t)')
plt.axvline(x=3, color='yellow', label='4δ(3t-9)')

plt.xlim(-4, 4)

plt.xlabel('time')
plt.title('Signal 4δ(3t-9) and δ(t)')
plt.legend()
plt.grid(True)
```
