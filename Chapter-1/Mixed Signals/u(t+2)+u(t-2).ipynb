```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
In this file I plot u(t+2)+u(t-2) step by step


### Step 1: u(t)

```python
t = np.linspace(-2, 100, 1000)
u = np.heaviside(t, 0)

plt.plot(t, u, 'red', linewidth=3)
plt.xlim(-4, 4)
plt.ylim(0, 2)

plt.title('Unit Time Step Signal')
plt.xlabel('time')
plt.ylabel('u(t)')
plt.style.use('dark_background')
plt.grid(True)
```

### Step 2: u(t+2)

```python
t = np.linspace(-2, 100, 1000)
f1 = np.heaviside(t + 2, 0)

plt.plot(t, f1, 'green', linewidth=3)
plt.xlim(-4, 4)
plt.ylim(0, 2)

plt.title('u(t+2)')
plt.xlabel('time')
plt.ylabel('x1(t)')
plt.style.use('dark_background')
plt.grid(True)
```

### Step 3: u(t+2)

```python
t = np.linspace(-2, 100, 1000)
f2 = np.heaviside(t - 2, 0)

plt.plot(t, f2, 'purple', linewidth=3)
plt.xlim(-4, 4)
plt.ylim(0, 2)

plt.title('u(t-2)')
plt.xlabel('time')
plt.ylabel('x2(t)')
plt.style.use('dark_background')
plt.grid(True)
```

### Step 4: u(t+2) + u(t-2)

```python
t = np.linspace(-2, 100, 1000)
f1 = np.heaviside(t + 2, 0)
f2 = np.heaviside(t - 2, 0)
x_t = f1 + f2

plt.plot(t, x_t, 'orange', linewidth=3)
plt.xlim(-3, 3)
plt.ylim(-3, 3)

plt.title('u(t+2)+u(t-2)')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.style.use('dark_background')
plt.grid(True)
```
