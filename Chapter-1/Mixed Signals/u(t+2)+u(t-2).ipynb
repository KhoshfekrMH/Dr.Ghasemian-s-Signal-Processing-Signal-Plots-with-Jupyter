```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
$$Signal: x(t) = u(t+2)+u(t-2)$$
- Range: t ∈ [-4, 100], y ∈ [0, 4], x ∈ [-4, 6]
- In four step:
  1. $u(t)$
  2. $x1(t) = u(t+2)$
  3. $x2(t) = u(t-2)$
  4. $x(t) = u(t+2)+u(t-2)$

### Step 1: $u(t)$

```python
t = np.linspace(-4, 100, 10000)
u = np.heaviside(t, 0)

plt.plot(t, u, 'red', linewidth=3, label='u(t)')

plt.xlim(-4, 6)
plt.ylim(0, 2)

plt.title('Unit Time Step Signal')
plt.xlabel('time')
plt.ylabel('u(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

### Step 2: $x1(t) = u(t+2)$

```python
t = np.linspace(-4, 100, 10000)
x1 = np.heaviside(t + 2, 0)

plt.plot(t, x1, 'aqua', linewidth=3, label='u(t+2)')

plt.xlim(-4, 6)
plt.ylim(0, 2)

plt.title('Step2: u(t+2)')
plt.xlabel('time')
plt.ylabel('x1(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

### Step 3: $u(t-2)$

```python
t = np.linspace(-4, 100, 10000)
x2 = np.heaviside(t - 2, 0)

plt.plot(t, x2, 'Orange', linewidth=3, label='u(t-2)')

plt.xlim(-4, 6)
plt.ylim(0, 2)

plt.title('Step3: u(t-2)')
plt.xlabel('time')
plt.ylabel('x2(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

### Step 4: u(t+2) + u(t-2)

```python
t = np.linspace(-4, 100, 10000)
# Creation of x1(t):
x1 = np.heaviside(t + 2, 0)
# Creation of x2(t):
x2 = np.heaviside(t - 2, 0)
# Creation of x(t):
x_t = x1 + x2

plt.plot(t, x_t, '#16FF00', linewidth=3, label='x(t)') 
plt.xlim(-3, 3)
plt.ylim(-3, 3)

plt.title('u(t+2)+u(t-2)')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

### Comparison

```python
t = np.linspace(-4, 100, 10000)
# Creation of x1(t):
x1 = np.heaviside(t + 2, 0)
# Creation of x2(t):
x2 = np.heaviside(t - 2, 0)
# Creation of x(t):
x_t = x1 + x2

plt.plot(t, x1, 'aqua', linewidth=3, label='x1(t)')
plt.plot(t, x2, 'orange', linewidth=3, label='x2(t)') 
plt.plot(t, x_t, '#16FF00', linewidth=6, alpha=0.3, label='x(t)')
plt.xlim(-4, 6)
plt.ylim(-3, 3)

plt.title('u(t+2)+u(t-2)')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```
