```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
$$Signal: x(t) = u(t)+u(t-1)+u(t-2)-3u(t-3)$$
- Range: t ∈ [-4, 100], y ∈ [0, 4], x ∈ [-4, 6]
- we could make a plot with 4 subplots:
1. $x1(t) = u(t)$
2. $x2(t) = u(t-1)$
3. $x3(t) = u(t-2)$
4. $x4(t) = 3u(t-3)$
- And then we could plot x1(t)+x2(t)+x3(t)-x4(t)

#### Step 1: $x1(t) = u(t)$

```python
t = np.linspace(-4, 100, 10000)
u = np.heaviside(t, 0) # Unit time step signal

plt.plot(t, u, 'red', linewidth=3, label='u(t)')

plt.xlim(-4, 6)
plt.ylim(0, 4)

plt.title('Step 1: u(t)')
plt.xlabel('time')
plt.ylabel('x1(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

#### Step 2: $x2(t) = u(t-1)$

```python
t = np.linspace(-4, 100, 10000)
x2 = np.heaviside(t - 1, 0) # u(t-1)

plt.plot(t, x2, 'yellow', linewidth=3, label='u(t-1)')

plt.xlim(-4, 6)
plt.ylim(0, 4)

plt.title('Step 2: u(t-1)')
plt.xlabel('time')
plt.ylabel('x2(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

#### Step 3: $x3(t) = u(t-2)$

```python
t = np.linspace(-4, 100, 10000)
x3 = np.heaviside(t - 2, 0) # u(t-2)

plt.plot(t, x3, 'aqua', linewidth=3, label='u(t-2)')

plt.xlim(-4, 6)
plt.ylim(0, 4)

plt.title('Step 2: u(t-2)')
plt.xlabel('time')
plt.ylabel('x3(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

#### Step 4: $x4(t) = 3u(t-3)$

```python
t = np.linspace(-4, 100, 10000)
f1 = np.heaviside(t - 3, 0) # u(t-3)
x4 = 3 * f1 # 3u(t-3)

plt.plot(t, x4, 'orange', linewidth=3, label='3u(t-3)')

plt.xlim(-4, 6)
plt.ylim(0, 4)

plt.title('Step 2: 3u(t-3)')
plt.xlabel('time')
plt.ylabel('x4(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

#### Step 5: $x(t) = x1(t)+x2(t)+x3(t)-x4(t)$

```python
t = np.linspace(-4, 100, 10000)
# creation x1(t):
x1 = np.heaviside(t, 0) # u(t)
# creation x2(t):
x2 = np.heaviside(t - 1, 0) # u(t-1)
# creation x3(t):
x3 = np.heaviside(t - 2, 0) # u(t-2)
# creation x4(t):
f1 = np.heaviside(t - 3, 0) # u(t-3)
x4 = 3 * f1 # 3u(t-3)
# creation x(t):
x_t = x1 + x2 + x3 - x4

plt.plot(t, x_t, '#F72798', linewidth=3, label='x(t)')

plt.xlim(-4, 6)
plt.ylim(0, 4)

plt.title('Step 5: u(t)+u(t-1)+u(t-2)-3u(t-3)')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```

#### Comparison

```python
t = np.linspace(-4, 100, 10000)
# creation x1(t):
x1 = np.heaviside(t, 0) # u(t)
# creation x2(t):
x2 = np.heaviside(t - 1, 0) # u(t-1)
# creation x3(t):
x3 = np.heaviside(t - 2, 0) # u(t-2)
# creation x4(t):
f1 = np.heaviside(t - 3, 0) # u(t-3)
x4 = 3 * f1 # 3u(t-3)
# creation x(t):
x_t = x1 + x2 + x3 - x4

plt.plot(t, x1, 'red', linewidth=2, label='u(t)') # u(t)
plt.plot(t, x2, 'yellow', linewidth=2, label='u(t-1)') # u(t-1)
plt.plot(t, x3, 'aqua', linewidth=2, label='u(t-2)') # u(t-2)
plt.plot(t, x4, 'orange', linewidth=2, label='3u(t-3)') # 3u(t-3)
plt.plot(t, x_t, '#F72798', linewidth=6, alpha=0.6, label='x(t)') # u(t)+u(t-1)+u(t-2)-3u(t-3)

plt.xlim(-4, 6)
plt.ylim(0, 4)

plt.title('Step 5: u(t)+u(t-1)+u(t-2)-3u(t-3)')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```
