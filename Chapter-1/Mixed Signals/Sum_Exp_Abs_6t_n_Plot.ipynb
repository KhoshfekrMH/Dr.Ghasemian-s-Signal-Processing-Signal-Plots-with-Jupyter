### Introduction

$$Signal: x(t) = \sum_{n=-\infty}^{\infty} e^{-|6t+n|}$$
- Range: t ∈ [1, 100] and n ∈ [-1000, 1000]
- However in this file we limit the range of n to [-1000, 1000] so mathematic formula is different:
<br>
$x(t) = \sum_{n=-1000}^{1000} e^{-|6t+n|}$
<br>
- The main reason that I did this because I could not plot this sequence in infinite range. 
- Also I limit time to 100s (every 6s).

<hr>

```python
import numpy as np
import matplotlib.pyplot as plt

t_values = np.arange(1, 100, 6)

n_values = np.arange(-1000, 1001)

plt.figure(figsize=(12, 8))

for t in t_values:
    result = np.exp(-np.abs(6*t + n_values))
    plt.plot(n_values, result, label=f't={t}')

plt.xlabel('n')
plt.ylabel('Value')
plt.title('Plot of Exp[-|6t+n|] for t in the range 1s to 100s (every 6s) and n in the range -1000 to 1000')
plt.legend()
plt.style.use('dark_background')
plt.grid(True)
```
