```python
import numpy as np
import matplotlib.pyplot as plt
```

### Introduction
In this file I plot u(t-)-u(t-1)+t[u(t-1)-u(t-2)]+2[u(t-2)-u(t-4)] step by step
- first of all we could make tree separated signals and then mix them together:
  1. x1(t) =  u(t) - u(t-1)
  2. x2(t) = t[u(t-1) - u(t-2)]
  3. x3(t) = 2[u(t-2) - u(t-4)]
  4. x(t) = u1 + u2 + u3

### Step 1(I): u(t)

```python
t = np.linspace(-2, 100, 1000)
u = np.heaviside(t, 0) # Unit time step signal

plt.plot(t, u, 'red', linewidth=3)
plt.xlim(-4, 4)
plt.ylim(0, 2)

plt.title('Step 1(I): u(t)')
plt.xlabel('time')
plt.ylabel('u(t)')
plt.style.use('dark_background')
plt.grid(True)
```

### Step 1(II): u(t-1)

```python
t = np.linspace(-2, 100, 1000)
f1 = np.heaviside(t - 1, 0) # u(t-1)

plt.plot(t, f1, 'red', linewidth=3)
plt.xlim(-4, 4)
plt.ylim(0, 2)

plt.title('Step 1(II): u(t-1)')
plt.xlabel('time')
plt.ylabel('f1(t)')
plt.style.use('dark_background')
plt.grid(True)
```

### Step 1(III): u1 = u(t) - u(t-1)

```python
t = np.linspace(-2, 100, 1000)
u = np.heaviside(t, 0) # u(t)
f1 = np.heaviside(t - 1, 0) # u(t-1)
x1 = u - f1 # u(t) - u(t-1)

plt.plot(t, x1, 'red', linewidth=3)
plt.xlim(-4, 4)
plt.ylim(0, 2)

plt.title('Step 1(III): x1 = u(t) - u(t-1)')
plt.xlabel('time')
plt.ylabel('x1(t)')
plt.style.use('dark_background')
plt.grid(True)
```

<hr>

### Step 2(I): u(t-2)

```python
t = np.linspace(-2, 100, 1000)
f1 = np.heaviside(t - 2, 0) # u(t-2)

plt.plot(t, f1, 'aqua', linewidth=3)
plt.xlim(-4, 4)
plt.ylim(0, 2)

plt.title('Step 2(I): u(t-2)')
plt.xlabel('time')
plt.ylabel('f1(t)')
plt.style.use('dark_background')
plt.grid(True)
```

### Step 2(II): u(t-1) - u(t-2)

```python
t = np.linspace(-2, 100, 1000)
f1 = np.heaviside(t - 1, 0) # u(t-1)
f2 = np.heaviside(t - 2, 0) # u(t-2)
f3 = f1 - f2 # u(t-1) - u(t-2)

plt.plot(t, f3, 'aqua', linewidth=3)
plt.xlim(-4, 4)
plt.ylim(0, 2)

plt.title('Step 2(II): u(t-1) - u(t-2)')
plt.xlabel('time')
plt.ylabel('f3(t)')
plt.style.use('dark_background')
plt.grid(True)
```

### Step 2(III): u2 = t[u(t-1) - u(t-2)]

```python
t = np.linspace(-2, 100, 1000)
f1 = np.heaviside(t - 1, 0) # u(t-1)
f2 = np.heaviside(t - 2, 0) # u(t-2)
sum = f1 - f2 # u(t-1) - u(t-2)
x2 = t * sum # t[u(t-1) - u(t-2)]

plt.plot(t, x2, 'aqua', linewidth=3)
plt.xlim(-4, 4)
plt.ylim(0, 2)

plt.title('Step 2(III): x2 = t[u(t-1) - u(t-2)]')
plt.xlabel('time')
plt.ylabel('x2(t)')
plt.style.use('dark_background')
plt.grid(True)
```

<hr>

### Step 3(I): u(t-4)

```python
t = np.linspace(-2, 100, 1000)
f1 = np.heaviside(t - 4, 0) # u(t-4)

plt.plot(t, f1, 'yellow', linewidth=3)
plt.xlim(-4, 6)
plt.ylim(0, 2)

plt.title('Step 3(I): u(t-4)')
plt.xlabel('time')
plt.ylabel('f1(t)')
plt.style.use('dark_background')
plt.grid(True)
```

#### Step 3(II): u(t-2) - u(t-4)

```python
t = np.linspace(-2, 100, 1000)
f1 = np.heaviside(t - 2, 0) # u(t-2)
f2 = np.heaviside(t - 4, 0) # u(t-4)
f3 = f1 - f2 # u(t-2) - u(t-4)

plt.plot(t, f3, 'yellow', linewidth=3)
plt.xlim(-4, 6)
plt.ylim(0, 2)

plt.title('Step 3(II): u(t-2) - u(t-4)')
plt.xlabel('time')
plt.ylabel('f3(t)')
plt.style.use('dark_background')
plt.grid(True)
```

#### Step 3(III): u3 = 2[u(t-2) - u(t-4)]

```python
t = np.linspace(-2, 100, 1000)
f1 = np.heaviside(t - 2, 0) # u(t-2)
f2 = np.heaviside(t - 4, 0) # u(t-4)
sum = f1 - f2 # u(t-2) - u(t-4)
x3 = 2 * sum # 2[u(t-2) - u(t-4)]

plt.plot(t, x3, 'yellow', linewidth=3)
plt.xlim(-6, 6)
plt.ylim(0, 3)

plt.title('Step 3(III): x3 = 2[u(t-2) - u(t-4)]')
plt.xlabel('time')
plt.ylabel('x3(t)')
plt.style.use('dark_background')
plt.grid(True)
```

<hr>

### Final Step: u = u1 + u2 + u3

```python
t = np.linspace(-2, 100, 1000)
# creation of u1:
f1 = np.heaviside(t, 0) # u(t)
f2 = np.heaviside(t - 1, 0) # u(t-1)
x1 = f1 - f2 # u(t) - u(t-1)
# creation of u2:
l1 = np.heaviside(t - 1, 0) # u(t-1)
l2 = np.heaviside(t - 2, 0) # u(t-2)
sum1 = l1 - l2 # u(t-1) - u(t-2)
x2 = t * sum1 # t[u(t-1) - u(t-2)]
# creation of u3:
y1 = np.heaviside(t - 2, 0) # u(t-2)
y2 = np.heaviside(t - 4, 0) # u(t-4)
sum2 = y1 - y2 # u(t-2) - u(t-4)
x3 = 2 * sum2 # 2[u(t-2) - u(t-4)]
# creation of x(t):
x_t = x1 + x2 + x3

plt.plot(t, x_t, '#F57D1F', linewidth=3)
plt.xlim(-6, 6)
plt.ylim(0, 3)

plt.title('Final Step: x(t) = u1 + u2 + u3')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.style.use('dark_background')
plt.grid(True)
```

#### Comparison:

```python
t = np.linspace(-2, 100, 1000)
# creation of u1:
f1 = np.heaviside(t, 0) # u(t)
f2 = np.heaviside(t - 1, 0) # u(t-1)
x1 = f1 - f2 # u(t) - u(t-1)
# creation of u2:
l1 = np.heaviside(t - 1, 0) # u(t-1)
l2 = np.heaviside(t - 2, 0) # u(t-2)
sum1 = l1 - l2 # u(t-1) - u(t-2)
x2 = t * sum1 # t[u(t-1) - u(t-2)]
# creation of u3:
y1 = np.heaviside(t - 2, 0) # u(t-2)
y2 = np.heaviside(t - 4, 0) # u(t-4)
sum2 = y1 - y2 # u(t-2) - u(t-4)
x3 = 2 * sum2 # 2[u(t-2) - u(t-4)]
# creation of x(t):
x_t = x1 + x2 + x3

plt.plot(t, x1, 'red', linewidth=2) # u(t) - u(t-1): red
plt.plot(t, x2, 'aqua', linewidth=2) # t[u(t-1) - u(t-2)]: aqua
plt.plot(t, x3, 'yellow', linewidth=2) # 2[u(t-2) - u(t-4)]: yellow
plt.plot(t, x_t, '#F57D1F', linewidth=6, alpha=0.6) # x(t): #F57D1F (nearly orange)
plt.xlim(-6, 6)
plt.ylim(0, 3)

plt.title('Compare Signals')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.style.use('dark_background')
plt.grid(True)
```
