```python
import numpy as np
import matplotlib.pyplot as plt
```

Introduction
--

If we know we have LTI system and have our h[n], then find system respond to x[n]

```python
# Define the signals h[n] and x[n]
h = np.array([1, 2])  # h[0] = 1, h[1] = 2
x = np.array([0, 0, 1, 1])  # x[2] = 1, x[3] = 1

# Perform the convolution
y = np.convolve(x, h)

# Define the range for n
n_h = np.arange(len(h))  # n range for h[n]
n_x = np.arange(len(x))  # n range for x[n]
n_y = np.arange(len(y))  # n range for y[n]

# Plot the signals
plt.figure(figsize=(12, 6))

# Plot h[n]
plt.figure(figsize=(12, 6))
plt.subplot(3, 1, 1)
plt.stem(n_h, h, 'r', label='h[n]')
plt.title('h[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Plot x[n]
plt.subplot(3, 1, 2)
plt.stem(n_x, x, 'r', label='x[n]')
plt.title('x[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Plot y[n]
plt.subplot(3, 1, 3)
plt.stem(n_y, y, 'r', label='y[n]')
plt.title('y[n] = x[n] * h[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.style.use('dark_background')
plt.tight_layout()
```
