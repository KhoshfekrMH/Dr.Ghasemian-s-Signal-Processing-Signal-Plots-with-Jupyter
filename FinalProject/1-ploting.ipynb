Signal-1) $x_{1} = e ^ {t}.u(t) - e^ {-t}.u(t)$
--

#### $e^{t}$ and $e^{-t}$

```python
# Define the exponential function
def e(t):
  return np.exp(t)

def en(t):
  return np.exp(-t)

# Create time range
t = np.linspace(-5, 5, 10000)  # Adjust range and number of points as needed

# Calculate function values
y1 = e(t)
y2 = en(t)

# Plot the function
plt.plot(t, y1, linewidth=3, color='yellow', label='$e_{t}$')
plt.plot(t, y2, linewidth=3, color='aqua', label='$e_{-t}$')

# Add labels and title
plt.xlabel('time')
plt.ylabel('$e^{t}$' + ' and ' '$e^{-t}$')
plt.title('the exponential functions')
plt.legend()

# Show the plot
plt.style.use('dark_background')
plt.grid(True)
```

#### $u(t)$

```python
# Define the unit time step function
def u(t):
  return np.heaviside(t, 0)

# Create time range
t = np.linspace(-5, 5, 10000)  # Adjust range and number of points as needed

# Calculate function values
y = u(t)

# Plot the function
plt.plot(t, y, linewidth=3, color='purple', label='$u(t)$')

# Add labels and title
plt.xlabel('time')
plt.ylabel('$u(t)$')
plt.title('Unit Time Step Signal')
plt.legend()

# Show the plot
plt.style.use('dark_background')
plt.grid(True)
```

#### $x_{1}(t) = e^{t}.u(t) - e^{-t}.u(t)$

```python
# Define the x1 function
def x1(t):
  return np.exp(t) * np.heaviside(t, 0) - np.exp(-t) * np.heaviside(-t, 0)

# Create time range
t = np.linspace(-5, 5, 10000)  # Adjust range and number of points as needed

# Calculate function values
y = x1(t)

# Plot the function
plt.plot(t, y, linewidth=3, color='red', label='$x_{1}$')

# Add labels and title
plt.xlabel('t')
plt.ylabel('$x_{1}(t)$')
plt.title('Plot of $x_{1}(t) = e^{t}.u(t) - e^{-t}.u(t)$')
plt.legend()

# Show the plot
plt.style.use('dark_background')
plt.grid(True)
```

#### Creation of $x_{1}$

```python
# Define functions
def x1(t):
  return np.exp(t) * np.heaviside(t, 0) - np.exp(-t) * np.heaviside(-t, 0)

def u(t):
  return np.heaviside(t, 0)

def e(t):
  return np.exp(t)

def en(t):
  return np.exp(-t)

# Create time range
t = np.linspace(-5, 5, 10000)  # Adjust range and number of points as needed

# Calculate functions values
y1 = x1(t)
y2 = e(t)
y3 = en(t)
y4 = u(t)

# Plot functions
plt.plot(t, y1, color='red', label='$x_{1}$')
plt.plot(t, y2, linewidth=5, color='yellow', label='$e^{t}$', alpha=0.3)
plt.plot(t, y3, linewidth=5, color='aqua', label='$e^{t}$', alpha=0.3)
plt.plot(t, y4, linewidth=5, color='purple', label='$u(t)$', alpha=0.3)

# Add labels and title
plt.xlabel('time')
plt.title('Creation of $x_{1}$')
plt.legend()

# Show the plot
plt.style.use('dark_background')
plt.grid(True)
```

Signal-2) $x_{2} = u(t) - u(t-2)$
--

#### $u(t-2)$

```python
# Define the unit time step function with +2 shift
def shifted_u(t):
  return np.heaviside(t-2, 0)

# Create time range
t = np.linspace(-5, 5, 10000)  # Adjust range and number of points as needed

# Calculate function values
y = shifted_u(t)

# Plot the function
plt.plot(t, y, linewidth=3, color='purple', label='$u(t-2)$')

# Add labels and title
plt.xlabel('time')
plt.ylabel('$u(t-2)$')
plt.title('Unit Time Step Signal with +2 shift')
plt.legend()

# Show the plot
plt.style.use('dark_background')
plt.grid(True)
```

#### $x_{2} = u(t) - u(t-2)$

```python
# Define x2 function
def x2(t):
  return np.heaviside(t, 0) - np.heaviside(t-2, 0)

# Create time range
t = np.linspace(-5, 5, 10000)  # Adjust range and number of points as needed

# Calculate function values
y = x2(t)

# Plot the function
plt.plot(t, y, linewidth=3, color='red', label='$x_{2}$')

# Add labels and title
plt.xlabel('time')
plt.ylabel('$x_{2}(t)$')
plt.title('Plot of $x_{2}(t) = u(t) - u(t-2)$')
plt.legend()

# Show the plot
plt.style.use('dark_background')
plt.grid(True)
```

#### Creation of $x_{2}$

```python
# Define functions
def x2(t):
  return np.heaviside(t, 0) - np.heaviside(t-2, 0)

def u(t):
  return np.heaviside(t, 0)

def shifted_u(t):
  return np.heaviside(t-2, 0)

# Create time range
t = np.linspace(-5, 5, 10000)  # Adjust range and number of points as needed

# Calculate functions values
y1 = x2(t)
y2 = u(t)
y3 = shifted_u(t)

# Plot the function
plt.plot(t, y1, color='red', label='$x_{2}$')
plt.plot(t, y2, linewidth=3, color='yellow', label='$u(t)$', alpha=0.3)
plt.plot(t, y2, linewidth=3, color='aqua', label='$u(t-2)$', alpha=0.4)

# Add labels and title
plt.xlabel('time')
plt.title('Creation of $x_{2}$')
plt.legend()

# Show the plot
plt.style.use('dark_background')
plt.grid(True)
```

Signal-4) $x_{4} = 2 ^ {n} (u[n] - u[n-6])$
--

#### $2^{n}$

```python
def f(n):
    return (2.0 ** n)

# Set the range for n
n = np.arange(-1, 8)

# Compute the values of the function
x_values = f(n)

# Plot the discrete signal
plt.stem(n, x_values, 'r', label='$2^{n}$')
plt.title(r'$2^{n}$')
plt.xlabel('n')
plt.ylabel('y')
plt.legend()

plt.style.use('dark_background')
plt.grid(True)
```

#### $x_{4} = 2 ^ {n} (u[n] - u[n-6])$

```python
def x4(n):
  u_n = np.where(n >= 0, 1, 0)
  u_n_minus_6 = np.where(n >= 6, 1, 0)
  return (2.0 ** n) * (u_n - u_n_minus_6)

# Set the range for n
n = np.arange(-1, 8)

# Compute the values of the function
x_values = x4(n)

# Plot the discrete signal
plt.stem(n, x_values, 'r', label=r'$x_{4}$')
plt.title(r'$x_4[n] = 2^n (u[n] - u[n-6])$')
plt.xlabel('n')
plt.ylabel('$x_{4}[n]$')
plt.legend()

# Show the plot
plt.style.use('dark_background')
plt.grid(True)
```

Signal-5) $x_{5} = \sin(0.25 \pi)(u[n+4] - u[n-4])$
--

#### $x_{5} = \sin(0.25 \pi)(u[n+4] - u[n-4])$

```python
def x5(n):
  u_n_plus_4 = np.where(n >= -4, 1, 0)  # u[n+4] is 1 for n >= -4, otherwise 0
  u_n_minus_4 = np.where(n >= 4, 1, 0)  # u[n-4] is 1 for n >= 4, otherwise 0
  return np.sin(0.25 * np.pi) * (u_n_plus_4 - u_n_minus_4)

# Set the range for n
n = np.arange(-10, 10)

# Compute the values of the function
y_values = x5(n)

# Plot the discrete signal
plt.stem(n, y_values, 'r', label='$x_{5}$')
plt.title(r'$x_5[n] = \sin(0.25 \pi)(u[n+4] - u[n-4])$')
plt.xlabel('n')
plt.ylabel('$x_5[n]$')
plt.legend()

# Show the plot
plt.style.use('dark_background')
plt.grid(True)
```

Signal-6) $x_{6} = 2^{|n|}(u[\frac{n}{2}+3] - u[\frac{n}{2}-3])$
--

#### $u[\frac{n}{2}+3] - u[\frac{n}{2}-3]$

```python
def u_function(n):
  u_1 = np.where(n/2 >= -3, 1, 0) #u[n/2 + 3]
  u_2 = np.where(n/2 >= 3, 1, 0) #u[n/2 - 3]
  return (u_1) - (u_2)

# Set the range for n
n = np.arange(-10, 10)

# Compute the values of the function
y_values = u_function(n)

# Plot the discrete signal
plt.stem(n, y_values, 'r')
plt.title(r'$u[\frac{n}{2}+3] - u[\frac{n}{2}-3]$')
plt.xlabel('n')
plt.ylabel('y')

# Show the plot
plt.style.use('dark_background')
plt.grid(True)
```

#### $2^{|n|}$

```python
def f(n):
  return (2.0 ** np.abs(n))

# Set the range for n
n = np.arange(-1, 8)

# Compute the values of the function
x_values = f(n)

# Plot the discrete signal
plt.stem(n, x_values, 'r')
plt.title(r'$2^{|n|}$')
plt.xlabel('n')
plt.ylabel('y')

# Show the plot
plt.style.use('dark_background')
plt.grid(True)
```

#### $x_{6} = 2^{|n|}(u[\frac{n}{2}+3] - u[\frac{n}{2}-3])$

```python
def x6(n):
  u_1 = np.where(n/2 >= -3, 1, 0) #u[n/2 + 3]
  u_2 = np.where(n/2 >= 3, 1, 0) #u[n/2 - 3]
  f = 2.0 ** np.abs(n)
  return f*((u_1) - (u_2))

# Set the range for n
n = np.arange(-10, 10)

# Compute the values of the function
y_values = x6(n)

# Plot the discrete signal
plt.stem(n, y_values, 'r',  label='$x_{6}$')
plt.title(r'$x_{6}[n] = 2^{|n|}(u[\frac{n}{2}+3] - u[\frac{n}{2}-3])$')
plt.xlabel('n')
plt.ylabel('$x_{6}[n]$')
plt.legend()

# Show the plot
plt.style.use('dark_background')
plt.grid(True)
```
